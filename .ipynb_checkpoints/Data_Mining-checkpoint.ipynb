```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
df = pd.read_csv('DataTCC.csv')
df.head()
```

```python
df.info()
```

## Analise ExploratÃ³ria de Dados

```python
df.isnull().sum()
```

Umas colunas devem ser retiradas pela quantidade de valores vazios

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
#Verificando se tem mais linhas duplicadas
df.duplicated().sum()
```

```python
df.columns
```

```python
#Excluindo objetos
df.drop(['data_iniSE','versao_modelo'],axis=1, inplace=True)
#Excluindo colunas nulas
df.drop(['umidmed', 'umidmin',
       'tempmed', 'tempmax', 'casprov', 'casprov_est', 'casprov_est_min',
       'casprov_est_max', 'casconf',],axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
#Colocando o resto dos valores nulos em 0
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
#Retirando colunas que iram ocasionar vicios
df.drop(['notif_accum_year','p_rt1','nivel','casos_est','casos_est_min','casos_est_max'],axis=1, inplace=True)
```

```python
df.columns
```

```python
df.corr()
```

```python
correlation = df.corr()
```

```python
# plot da matriz de correlaÃ§Ã£o

plot = sn.heatmap(correlation, annot = True, fmt=".1f", linewidths=.6)
plot
```

```python
#Retirando colunas com pouca correlaÃ§Ã£o
df.drop(['Localidade_id','id','IBGE'],axis=1, inplace=True)
```

## Normalizando

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
df_sc = scale.fit_transform(df)
df_sc = pd.DataFrame(df_sc, columns=df.columns)
```

```python
df_sc.head()
```

# Definindo variÃ¡veis e dados

```python
y = df_sc['casos']
X = df_sc.drop('casos', axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

# SVR

```python
from sklearn.svm import SVR
```

```python
# Ajustando o SVR ao conjunto de treinamento
SVM_regression = SVR(kernel = 'rbf', C=10, gamma=0.01)
SVM_regression.fit(X_train, y_train)
```

```python
# Prevendo os resultados do conjunto de teste
y_hat = SVM_regression.predict(X_test)
```

```python
predictions = pd.DataFrame({ 'y_test':y_test,'y_hat':y_hat})
predictions.head()
```

```python
SVM_regression.score(X_test, y_test)
```

```python
sn.scatterplot(x=y_test, y=y_hat, alpha=0.6)
sn.lineplot(y_test, y_test)

plt.xlabel('Quantidade de casos atual', fontsize=14)
plt.ylabel('Quantidade de casos previstos', fontsize=14)
plt.title('Atual vs Casos previstos (test set)', fontsize=17)
plt.show()
```

```python
MSE_test = round(np.mean(np.square(y_test - y_hat)),2)
RMSE_test = round(np.sqrt(MSE_test),2)
RMSE_test
```

```python
MSE_test
```

### Ajustando hiperparÃ¢metros:

```python
my_param_grid = {'C': [1,10,100], 'gamma': [1,0.1,0.01], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
GridSearchCV(estimator=SVR(),param_grid= my_param_grid, refit = True, verbose=3, cv=5 )
```

```python
grid = GridSearchCV(estimator=SVR(),param_grid= my_param_grid, refit = True, verbose=2, cv=2 )
#verbose significa apenas a saÃ­da de texto que descreve o processo. (quanto maior o nÃºmero, mais detalhes vocÃª obterÃ¡).
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
#irÃ¡ pegar os melhores parametros
```

```python
grid.best_estimator_
```

```python
y_hat_optimized = grid.predict(X_test)
```

```python
predictions['y_hat_optimized'] = y_hat_optimized
predictions.head()
```

```python
sn.scatterplot(x=y_test, y=y_hat, alpha=0.6)
sn.lineplot(y_test, y_test)

plt.xlabel('Quantidade de casos atual', fontsize=14)
plt.ylabel('Quantidade de casos previstos', fontsize=14)
plt.title('Atual vs Casos previstos (test set)', fontsize=17)
plt.show()
```

```python
SVR_score_opt = grid.score(X_test, y_test)
```

```python
SVR_score_opt
```

```python
MSE_SVR_opt = round(np.mean(np.square(y_test - y_hat_optimized)),2)
RMSE_SVR_opt = round(np.sqrt(MSE_SVR_opt),2)
RMSE_SVR_opt
```

```python
MSE_SVR_opt
```

# KNN - Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Ajustando a regressÃ£o de KNN ao conjunto de treinamento

KNN_regression = KNeighborsRegressor(n_neighbors=2)
KNN_regression.fit(X_train, y_train)
```

```python
y_hat = KNN_regression.predict(X_test)
```

```python
KNN_regression.score(X_test, y_test)
```

```python
sn.scatterplot(x=y_test, y=y_hat, alpha=0.6)
sn.lineplot(y_test, y_test)

plt.xlabel('Quantidade de casos atual', fontsize=14)
plt.ylabel('Quantidade de casos previstos', fontsize=14)
plt.title('Atual vs Casos previstos (conjunto de teste)', fontsize=17)
plt.show()
```

```python
MSE_KNN_test = round(np.mean(np.square(y_test - y_hat)),2)
RMSE_KNN_test = round(np.sqrt(MSE_KNN_test),2)
RMSE_KNN_test
```

```python
MSE_KNN_test
```

### Escolhendo K

```python
RMSE_KNN_CH_test = []
k=40
```

```python
for i in range(1,k):
    KNN_i = KNeighborsRegressor(n_neighbors=i)
    KNN_i.fit(X_train, y_train)
    RMSE_KNN_CH_test.append(np.sqrt(np.mean(np.square(y_test - KNN_i.predict(X_test)))))
    
optimal_k = pd.DataFrame({'RMSE_test':np.round(RMSE_KNN_CH_test,2)}, index=range(1,k))
```

```python
optimal_k.head(10)
```

```python
plt.figure(figsize=(10,5))
sn.lineplot(data=optimal_k)
plt.title('Cross Validated RMSE VS K')
plt.xlabel('K')
plt.ylabel('RMSE')
plt.show()
```

## ValidaÃ§Ã£o Cruzada

```python
from sklearn.model_selection import cross_val_score
```

### SVR

```python
cvsScoreSVR = max(cross_val_score(estimator = SVR(C=10, gamma=0.01), X = X_train, y = y_train, cv = 5))
```

```python
cvsScoreSVR
```

```python
NMSE = cross_val_score(estimator = SVR(C=10, gamma=0.01), X = X_train, y = y_train, cv = 5 , scoring="neg_mean_squared_error" )
MSE_CV_SVR = round(np.mean(-NMSE),4)
RMSE_CV_SVR = round(np.sqrt(MSE_CV_SVR),4)
RMSE_CV_SVR
```

```python
MSE_CV_SVR
```

### Random Forest

```python
cvsScoreRF = max(cross_val_score(estimator = RandomForestRegressor(max_depth=20, max_features='sqrt', n_estimators=500), X = X_train, y = y_train, cv = 5))
```

```python
cvsScoreRF
```

```python
NMSE = cross_val_score(estimator = RandomForestRegressor(max_depth=20, max_features='sqrt', n_estimators=500), X = X_train, y = y_train, cv = 5, scoring="neg_mean_squared_error")
MSE_CV_RF = round(np.mean(-NMSE),4)
RMSE_CV_RF = round(np.sqrt(MSE_CV_RF),4)
RMSE_CV_RF
```

```python
MSE_CV_RF
```

### KNN

```python
cvsScoreKNN = max(cross_val_score(estimator = KNN_regression, X = X_train, y = y_train, cv = 5 ))
```

```python
cvsScoreKNN
```

```python
NMSE = cross_val_score(estimator = KNN_regression, X = X_train, y = y_train, cv = 5 , scoring="neg_mean_squared_error" )
MSE_CV_KNN = round(np.mean(-NMSE),4)
RMSE_CV_KNN = round(np.sqrt(MSE_CV_KNN),4)
RMSE_CV_KNN
```

```python
MSE_CV_KNN
```
