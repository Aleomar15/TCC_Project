```python
import pandas as pd
import numpy as np
```

```python
url = "https://info.dengue.mat.br/api/alertcity"
ibge = list(map(lambda x: int(x.strip('[]')), listIBGE[0]))
geocode = ibge[0]
disease = "dengue"
format = "csv"
ew_start = 1
ew_end = 53
ey_start = 2017
ey_end = 2022

params =(
     "&disease="
     + f"{disease}"
     + "&geocode="
     + f"{geocode}"
     + "&disease="
     + f"{disease}"
    + "&format="
    + f"{format}"
    + "&ew_start="
    + f"{ew_start}"
    + "&ew_end="
    + f"{ew_end}"
    + "&ey_start=" 
    + f"{ey_start}"
    + "&ey_end="
    + f"{ey_end}"    
)

url_resp = "?".join([url, params])
dados = pd.read_csv(url_resp, index_col='SE')
dados['IBGE'] = ibge[0]
dados.T
```

```python
for elemento in listIBGE:
    ibge = list(map(lambda x: int(x.strip('[]')), elemento))
    ibge2 = elemento[0]
    url = "https://info.dengue.mat.br/api/alertcity"
    geocode = ibge2
    disease = "dengue"
    format = "csv"
    ew_start = 1
    ew_end = 53
    ey_start = 2017
    ey_end = 2022

    params =(
        "&disease="
        + f"{disease}"
        + "&geocode="
        + f"{geocode}"
        + "&disease="
        + f"{disease}"
        + "&format="
        + f"{format}"
        + "&ew_start="
        + f"{ew_start}"
        + "&ew_end="
        + f"{ew_end}"
        + "&ey_start=" 
        + f"{ey_start}"
        + "&ey_end="
        + f"{ey_end}"    
    )

    url_resp = "?".join([url, params])
    dadosAD = pd.read_csv(url_resp, index_col='SE')
    dadosAD ['IBGE'] = ibge[0]
    df_concatenado = pd.concat([dados, dadosAD])
    dados = df_concatenado
```
